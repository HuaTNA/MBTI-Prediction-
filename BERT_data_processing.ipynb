```python
# CNN-MBTI Classifier
# Using CNN model for MBTI personality type prediction

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import re
from collections import Counter
import time
from tqdm import tqdm
import nltk
from nltk.corpus import stopwords

# Set random seed for reproducibility
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# Check GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Download necessary NLTK resources
try:
    nltk.download('stopwords')
    print("NLTK resources downloaded successfully")
except:
    print("NLTK resources already exist or download failed")

# 1. Text preprocessing function
def preprocess_text(text):
    """Clean text data"""
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Keep only letters and spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text)
    # Convert to lowercase and strip whitespace
    text = text.lower().strip()
    return text

# 2. Vocabulary building function
def build_vocab(texts, max_vocab_size=30000, min_freq=3):
    """Build vocabulary from texts"""
    word_counts = Counter()

    # Count word frequencies
    for text in texts:
        words = text.split()
        word_counts.update(words)

    # Filter low-frequency words
    filtered_words = {word: count for word, count in word_counts.items() if count >= min_freq}

    # Select most common words for vocabulary
    vocab_words = sorted(filtered_words.items(), key=lambda x: x[1], reverse=True)[:max_vocab_size-2]  # Reserve 2 positions for <PAD> and <UNK>

    # Create word-to-index mapping
    word_to_idx = {'<PAD>': 0, '<UNK>': 1}
    word_to_idx.update({word: idx+2 for idx, (word, _) in enumerate(vocab_words)})

    return word_to_idx

# 3. Convert text to indices function
def text_to_indices(text, word_to_idx, max_len=500):
    """Convert text to index sequence"""
    words = text.split()
    indices = [word_to_idx.get(word, word_to_idx['<UNK>']) for word in words[:max_len]]

    # Pad or truncate to fixed length
    if len(indices) < max_len:
        indices += [word_to_idx['<PAD>']] * (max_len - len(indices))
    else:
        indices = indices[:max_len]

    return indices

# 4. Custom MBTI dataset class
class MBTIDataset(Dataset):
    """MBTI dataset class"""
    def __init__(self, texts, labels, word_to_idx, max_len=500):
        self.texts = texts
        self.labels = labels
        self.word_to_idx = word_to_idx
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]

        # Convert text to index sequence
        indices = text_to_indices(text, self.word_to_idx, self.max_len)

        return {
            'text': text,
            'indices': torch.tensor(indices, dtype=torch.long),
            'label': torch.tensor(label, dtype=torch.long)
        }

# 5. Data preparation function
def prepare_data(df, max_vocab_size=30000, max_len=500, augment=True):
    """Prepare data for CNN model"""
    print("Starting data preprocessing...")

    # Check if processed columns already exist, otherwise process
    if 'combined_clean_posts' not in df.columns:
        # Split posts
        print("Splitting posts...")
        if not isinstance(df["posts"].iloc[0], list):
            df["posts"] = df["posts"].apply(lambda x: x.split("|||"))

        # Clean text
        print("Cleaning text...")
        df["clean_posts"] = df["posts"].apply(lambda posts: [preprocess_text(p) for p in posts])

        # Combine cleaned posts into a single text
        print("Combining posts...")
        df["combined_clean_posts"] = df["clean_posts"].apply(lambda posts: " ".join(posts))

    # Truncate long texts
    print(f"Truncating text to max length of {max_len*10} characters...")
    df["truncated_text"] = df["combined_clean_posts"].apply(lambda x: x[:max_len*10])

    # Encode labels
    print("Encoding MBTI types...")
    label_encoder = LabelEncoder()
    df["label"] = label_encoder.fit_transform(df["type"])

    # Display label mapping and distribution
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    print(f"Label mapping: {label_mapping}")
    label_counts = Counter(df["label"])
    print(f"Label distribution: {label_counts}")

    # Split train and test sets
    train_df, test_df = train_test_split(
        df, test_size=0.2, random_state=SEED, stratify=df["label"]
    )

    # Data augmentation - duplicate samples for minority classes
    if augment:
        print("Applying data augmentation...")
        # Calculate target count for each class
        class_counts = Counter(train_df["label"])
        max_count = max(class_counts.values())
        target_count = int(max_count * 0.5)  # At least 50% of largest class

        augmented_dfs = []

        # Oversample minority classes
        for label, count in class_counts.items():
            if count < target_count:
                # Get samples for this class
                class_samples = train_df[train_df["label"] == label]
                # Number of duplications needed
                multiplier = target_count // count + 1
                # Duplicate samples
                for _ in range(multiplier):
                    if len(augmented_dfs) * count + len(class_samples) <= target_count:
                        augmented_dfs.append(class_samples)

        # Merge augmented data
        if augmented_dfs:
            augment_df = pd.concat(augmented_dfs, ignore_index=True)
            train_df = pd.concat([train_df, augment_df], ignore_index=True)
            print(f"Added {len(augment_df)} samples through data augmentation")
            print(f"Augmented training set distribution: {Counter(train_df['label'])}")

    # Build vocabulary
    print("Building vocabulary...")
    word_to_idx = build_vocab(train_df["truncated_text"].values, max_vocab_size=max_vocab_size)
    vocab_size = len(word_to_idx)
    print(f"Vocabulary size: {vocab_size}")

    # Create datasets
    train_dataset = MBTIDataset(
        texts=train_df["truncated_text"].values,
        labels=train_df["label"].values,
        word_to_idx=word_to_idx,
        max_len=max_len
    )

    test_dataset = MBTIDataset(
        texts=test_df["truncated_text"].values,
        labels=test_df["label"].values,
        word_to_idx=word_to_idx,
        max_len=max_len
    )

    print(f"Training set size: {len(train_dataset)}, Test set size: {len(test_dataset)}")

    return train_dataset, test_dataset, label_encoder, word_to_idx, vocab_size

# 6. CNN classification model
class CNNMBTIClassifier(nn.Module):
    """MBTI classifier using CNN"""
    def __init__(self, vocab_size, embedding_dim=300, num_filters=128, filter_sizes=[3, 4, 5],
                 hidden_dim=128, num_classes=16, dropout=0.5, max_len=500):
        super(CNNMBTIClassifier, self).__init__()

        # Embedding layer
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)

        # Convolutional layers
        self.convs = nn.ModuleList([
            nn.Conv1d(embedding_dim, num_filters, kernel_size=fs)
            for fs in filter_sizes
        ])

        # Fully connected layers and dropout
        self.fc1 = nn.Linear(len(filter_sizes) * num_filters, hidden_dim)
        self.bn = nn.BatchNorm1d(hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, num_classes)
        self.dropout = nn.Dropout(dropout)
        self.relu = nn.ReLU()

    def forward(self, x):
        # Input x shape: (batch_size, max_len)

        # Embedding layer
        embedded = self.embedding(x)  # (batch_size, max_len, embedding_dim)

        # Format embeddings for convolutional layer
        embedded = embedded.permute(0, 2, 1)  # (batch_size, embedding_dim, max_len)

        # Apply convolution and pooling for each filter size
        conved = [self.relu(conv(embedded)) for conv in self.convs]

        # Global max pooling
        pooled = [torch.max(conv, dim=2)[0] for conv in conved]

        # Concatenate all features
        cat = torch.cat(pooled, dim=1)

        # Fully connected layers
        x = self.dropout(cat)
        x = self.fc1(x)
        x = self.bn(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)

        return x

# 7. Training function
def train_cnn_model(model, train_loader, val_loader, optimizer, criterion, epochs=10, early_stopping_patience=3):
    """Train CNN model"""
    best_val_loss = float('inf')
    best_val_acc = 0.0
    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []

    # Early stopping counter
    counter = 0

    for epoch in range(epochs):
        print(f'\n{"-" * 20} Epoch {epoch+1}/{epochs} {"-" * 20}')

        # Training mode
        model.train()
        train_loss = 0
        train_acc = 0

        # Training loop
        for batch in tqdm(train_loader, desc="Training"):
            # Get batch data
            indices = batch['indices'].to(device)
            labels = batch['label'].to(device)

            # Zero gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(indices)

            # Calculate loss
            loss = criterion(outputs, labels)

            # Backward pass
            loss.backward()

            # Update parameters
            optimizer.step()

            # Calculate accuracy
            _, preds = torch.max(outputs, dim=1)
            correct = (preds == labels).float().sum()
            acc = correct / len(labels)

            train_loss += loss.item()
            train_acc += acc.item()

        # Calculate average training loss and accuracy
        avg_train_loss = train_loss / len(train_loader)
        avg_train_acc = train_acc / len(train_loader)
        train_losses.append(avg_train_loss)
        train_accs.append(avg_train_acc)

        # Validation mode
        model.eval()
        val_loss = 0
        val_acc = 0

        with torch.no_grad():
            for batch in tqdm(val_loader, desc="Validation"):
                # Get batch data
                indices = batch['indices'].to(device)
                labels = batch['label'].to(device)

                # Forward pass
                outputs = model(indices)

                # Calculate loss
                loss = criterion(outputs, labels)

                # Calculate accuracy
                _, preds = torch.max(outputs, dim=1)
                correct = (preds == labels).float().sum()
                acc = correct / len(labels)

                val_loss += loss.item()
                val_acc += acc.item()

        # Calculate average validation loss and accuracy
        avg_val_loss = val_loss / len(val_loader)
        avg_val_acc = val_acc / len(val_loader)
        val_losses.append(avg_val_loss)
        val_accs.append(avg_val_acc)

        # Print results
        print(f'Train Loss: {avg_train_loss:.4f} | Train Acc: {avg_train_acc*100:.2f}%')
        print(f'Val. Loss: {avg_val_loss:.4f} | Val. Acc: {avg_val_acc*100:.2f}%')

        # Save best model
        if avg_val_acc > best_val_acc:
            best_val_acc = avg_val_acc
            torch.save(model.state_dict(), 'best-cnn-mbti-model.pt')
            print(f"[!] Saved best model (validation accuracy: {best_val_acc*100:.2f}%)")
            counter = 0  # Reset early stopping counter
        else:
            counter += 1
            print(f"Early stopping counter: {counter}/{early_stopping_patience}")

        # Early stopping
        if counter >= early_stopping_patience:
            print(f"Early stopping! No improvement in validation accuracy for {early_stopping_patience} epochs")
            break

    # Plot training progress
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label='Train Accuracy')
    plt.plot(val_accs, label='Val Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.savefig('cnn_mbti_training.png')
    plt.show()

    return train_losses, val_losses, train_accs, val_accs

# 8. Evaluation function
def evaluate_cnn_model(model, test_loader, label_encoder):
    """Evaluate CNN model performance"""
    model.eval()
    predictions = []
    actual_labels = []

    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Evaluating"):
            # Get batch data
            indices = batch['indices'].to(device)
            labels = batch['label'].to(device)

            # Forward pass
            outputs = model(indices)

            # Get predictions
            _, preds = torch.max(outputs, dim=1)

            predictions.extend(preds.cpu().tolist())
            actual_labels.extend(labels.cpu().tolist())

    # Calculate accuracy
    accuracy = accuracy_score(actual_labels, predictions)

    # Generate classification report
    class_names = label_encoder.classes_
    report = classification_report(actual_labels, predictions, target_names=class_names)

    # Confusion matrix
    plt.figure(figsize=(12, 10))
    cm = confusion_matrix(actual_labels, predictions)
    cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

    sns.heatmap(cm_percent, annot=True, fmt='.1f',
                xticklabels=class_names,
                yticklabels=class_names,
                cmap='Blues')
    plt.title('Confusion Matrix (Percent)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig('cnn_mbti_confusion_matrix.png')

    return accuracy, report

# 9. Prediction function
def predict_with_cnn(text, model, word_to_idx, label_encoder, max_len=500):
    """Predict MBTI type for a single text using CNN model"""
    model.eval()

    # Text preprocessing
    text = preprocess_text(text)

    # Convert text to index sequence
    indices = text_to_indices(text, word_to_idx, max_len)
    indices_tensor = torch.tensor([indices], dtype=torch.long).to(device)

    # Prediction
    with torch.no_grad():
        outputs = model(indices_tensor)

        # Get prediction and probabilities
        probabilities = torch.softmax(outputs, dim=1)[0]
        predicted_class = torch.argmax(probabilities).item()

    # Convert prediction to MBTI type
    predicted_mbti = label_encoder.inverse_transform([predicted_class])[0]

    # Get probabilities for all types
    all_probs = {label_encoder.inverse_transform([i])[0]: prob.item()
                for i, prob in enumerate(probabilities)}

    # Sort probabilities
    sorted_probs = {k: v for k, v in sorted(all_probs.items(), key=lambda item: item[1], reverse=True)}

    return predicted_mbti, sorted_probs

# 10. Analyze performance for each MBTI dimension
def analyze_mbti_dimensions(actual_labels, predictions, label_encoder):
    """Analyze classification performance for each MBTI dimension"""

    # Convert integer labels to MBTI types
    actual_types = label_encoder.inverse_transform(actual_labels)
    predicted_types = label_encoder.inverse_transform(predictions)

    # Initialize counters
    dimension_counts = {
        'I/E': {'correct': 0, 'total': 0},
        'N/S': {'correct': 0, 'total': 0},
        'T/F': {'correct': 0, 'total': 0},
        'J/P': {'correct': 0, 'total': 0}
    }

    # Count correct predictions for each dimension
    for actual, predicted in zip(actual_types, predicted_types):
        # I/E dimension
        if actual[0] == predicted[0]:
            dimension_counts['I/E']['correct'] += 1
        dimension_counts['I/E']['total'] += 1

        # N/S dimension
        if actual[1] == predicted[1]:
            dimension_counts['N/S']['correct'] += 1
        dimension_counts['N/S']['total'] += 1

        # T/F dimension
        if actual[2] == predicted[2]:
            dimension_counts['T/F']['correct'] += 1
        dimension_counts['T/F']['total'] += 1

        # J/P dimension
        if actual[3] == predicted[3]:
            dimension_counts['J/P']['correct'] += 1
        dimension_counts['J/P']['total'] += 1

    # Calculate accuracy for each dimension
    dimension_accuracy = {}
    for dim, counts in dimension_counts.items():
        accuracy = counts['correct'] / counts['total'] * 100
        dimension_accuracy[dim] = accuracy
        print(f"{dim} dimension accuracy: {accuracy:.2f}%")

    # Plot dimension accuracy
    plt.figure(figsize=(10, 6))
    dimensions = list(dimension_accuracy.keys())
    accuracies = list(dimension_accuracy.values())

    plt.bar(dimensions, accuracies, color='skyblue')
    plt.ylim(0, 100)
    plt.xlabel('MBTI Dimension')
    plt.ylabel('Accuracy (%)')
    plt.title('Accuracy by MBTI Dimension')

    for i, acc in enumerate(accuracies):
        plt.text(i, acc + 2, f'{acc:.1f}%', ha='center')

    plt.tight_layout()
    plt.savefig('mbti_dimension_accuracy.png')
    plt.show()

    return dimension_accuracy

# 11. Main function - Run the CNN model
def run_cnn_model(df, embedding_dim=300, max_vocab_size=30000, max_len=500,
                 batch_size=32, epochs=10, learning_rate=0.001):
    """Run CNN model for MBTI prediction"""
    print(f"Using CNN model for MBTI classification")

    # Prepare data
    train_dataset, test_dataset, label_encoder, word_to_idx, vocab_size = prepare_data(
        df, max_vocab_size=max_vocab_size, max_len=max_len
    )

    # Create weighted random sampler to handle class imbalance
    train_labels = [sample['label'].item() for sample in train_dataset]
    class_counts = Counter(train_labels)
    class_weights = {class_id: 1.0 / count for class_id, count in class_counts.items()}
    sample_weights = [class_weights[label] for label in train_labels]
    sampler = WeightedRandomSampler(
        weights=sample_weights,
        num_samples=len(train_dataset),
        replacement=True
    )

    # Create data loaders
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        sampler=sampler
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size
    )

    # Initialize model
    num_classes = len(label_encoder.classes_)
    model = CNNMBTIClassifier(
        vocab_size=vocab_size,
        embedding_dim=embedding_dim,
        num_filters=128,
        filter_sizes=[3, 4, 5],
        hidden_dim=128,
        num_classes=num_classes,
        dropout=0.5,
        max_len=max_len
    )
    model = model.to(device)

    # Print model architecture
    print(f"Model architecture:\n{model}")

    # Calculate model parameters
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Total model parameters: {total_params:,}")
    print(f"Trainable parameters: {trainable_params:,}")

    # Define optimizer and loss function
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # Calculate class weights
    class_weights = torch.FloatTensor([
        1.0 / (class_counts[i] / len(train_labels))
        for i in range(num_classes)
    ]).to(device)

    # Define loss function with class weights
    criterion = nn.CrossEntropyLoss(weight=class_weights)

    # Train model
    print("\nStarting CNN model training...")
    train_cnn_model(model, train_loader, test_loader, optimizer, criterion, epochs=epochs)

    # Load best model
    model.load_state_dict(torch.load('best-cnn-mbti-model.pt'))

    # Evaluate model
    print("\nEvaluating the best model...")
    accuracy, report = evaluate_cnn_model(model, test_loader, label_encoder)

    print(f"\nTest set accuracy: {accuracy*100:.2f}%")
    print(f"\nClassification report:\n{report}")

    # Get model predictions on test set
    model.eval()
    all_predictions = []
    all_labels = []

    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Getting predictions"):
            indices = batch['indices'].to(device)
            labels = batch['label'].to(device)
            outputs = model(indices)
            _, preds = torch.max(outputs, dim=1)
            all_predictions.extend(preds.cpu().tolist())
            all_labels.extend(labels.cpu().tolist())

    # Analyze MBTI dimension performance
    print("\nAnalyzing MBTI dimension performance:")
    dimension_accuracy = analyze_mbti_dimensions(all_labels, all_predictions, label_encoder)

    # Sample predictions
    test_texts = [
        "I enjoy spending time alone with books. I prefer planning everything in advance and following a schedule. I need time to process my emotions and often analyze why I feel a certain way.",
        "I love parties and meeting new people! I'm spontaneous and go with the flow. I make decisions based on how I feel and what aligns with my values.",
        "I'm analytical and logical. I enjoy solving complex problems and developing systems. I prefer working independently and dislike unnecessary social interactions."
    ]

    print("\nSample text predictions:")
    for i, text in enumerate(test_texts):
        mbti_type, probabilities = predict_with_cnn(text, model, word_to_idx, label_encoder, max_len)

        print(f"\nExample {i+1}: {text[:100]}...")
        print(f"Predicted MBTI type: {mbti_type}")
        print("Type probabilities:")
        for j, (mbti, prob) in enumerate(probabilities.items()):
            if j < 5:  # Only show top 5
                print(f"{mbti}: {prob:.4f}")

    return {
        'model': model,
        'word_to_idx': word_to_idx,
        'label_encoder': label_encoder,
        'accuracy': accuracy,
        'report': report,
        'dimension_accuracy': dimension_accuracy
    }

# Main program
if __name__ == "__main__":
    # Read dataset
    # Assume your data is stored in CSV format with 'type' and 'posts' columns
    # df = pd.read_csv('mbti_dataset.csv')

    # If you already have loaded df DataFrame, you can use it directly
    print(f"Dataset shape: {df.shape}")

    # Run CNN model
    results = run_cnn_model(
        df,
        embedding_dim=300,      # Word embedding dimension
        max_vocab_size=30000,   # Maximum vocabulary size
        max_len=500,            # Maximum sequence length
        batch_size=32,          # Batch size
        epochs=50,              # Training epochs
        learning_rate=0.001     # Learning rate
    )

    print("\nCNN model training and evaluation completed!")
```

```python
# MBTI性格类型分类 - 使用已加载的数据
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

# 下载必要的NLTK资源
try:
    nltk.download('stopwords')
    nltk.download('wordnet')
    print("NLTK资源下载完成")
except:
    print("NLTK资源已存在或下载失败")

# 1. 预处理函数
def preprocess_text(text):
    # 删除URL
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # 删除HTML标签
    text = re.sub(r'<.*?>', '', text)
    # 只保留字母和空格
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # 转换为小写
    text = text.lower().strip()
    # 删除多余空格
    text = re.sub(r'\s+', ' ', text)
    return text

# 2. 主要处理函数
def process_data(df):
    print("开始数据处理...")

    # 确保posts列存在
    if "posts" not in df.columns:
        print("错误：数据集中没有'posts'列")
        return None

    # 分割帖子
    print("分割帖子...")
    # 检查posts是否已经是列表
    if isinstance(df["posts"].iloc[0], list):
        print("帖子已经是列表格式")
    else:
        df["posts"] = df["posts"].apply(lambda x: x.split("|||"))

    # 清理文本
    print("清理文本...")
    df["clean_posts"] = df["posts"].apply(lambda posts: [preprocess_text(p) for p in posts])

    # 合并清理后的帖子为一个文本
    print("合并帖子...")
    df["combined_clean_posts"] = df["clean_posts"].apply(lambda posts: " ".join(posts))

    # 编码标签
    print("编码MBTI类型...")
    label_encoder = LabelEncoder()
    df["label"] = label_encoder.fit_transform(df["type"])
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    print(f"标签映射: {label_mapping}")

    # 检查标签分布
    label_counts = Counter(df["label"])
    print(f"标签分布: {label_counts}")

    return df, label_encoder

# 3. 特征提取与模型构建
def build_models(df, label_encoder):
    print("\n开始特征提取与模型构建...")

    # 分割数据
    print("分割训练集和测试集...")
    X_train_text, X_test_text, y_train, y_test = train_test_split(
        df["combined_clean_posts"],
        df["label"],
        test_size=0.2,
        random_state=42,
        stratify=df["label"]  # 确保各类别比例一致
    )

    # TF-IDF特征提取
    print("TF-IDF特征提取...")
    vectorizer = TfidfVectorizer(
        max_features=5000,  # 前5000个最重要的词
        min_df=5,           # 至少在5个文档中出现
        max_df=0.7,         # 最多在70%的文档中出现
        ngram_range=(1, 2)  # 包括单词和双词短语
    )

    X_train_tfidf = vectorizer.fit_transform(X_train_text)
    X_test_tfidf = vectorizer.transform(X_test_text)

    print(f"特征维度: {X_train_tfidf.shape[1]}")

    # 训练多个模型并比较
    print("\n训练基线模型...")

    # 1. 支持向量机模型 (带类别权重平衡)
    print("训练SVM模型...")
    svm_model = SVC(kernel="linear", C=1.0, class_weight='balanced')
    svm_model.fit(X_train_tfidf, y_train)

    # 预测与评估
    y_pred_svm = svm_model.predict(X_test_tfidf)
    svm_accuracy = accuracy_score(y_test, y_pred_svm)
    print(f"SVM准确率: {svm_accuracy:.4f}")

    # 2. 随机森林
    print("训练随机森林模型...")
    rf_model = RandomForestClassifier(
        n_estimators=100,
        max_depth=20,
        random_state=42,
        class_weight='balanced'
    )
    rf_model.fit(X_train_tfidf, y_train)

    y_pred_rf = rf_model.predict(X_test_tfidf)
    rf_accuracy = accuracy_score(y_test, y_pred_rf)
    print(f"随机森林准确率: {rf_accuracy:.4f}")

    # 3. 逻辑回归
    print("训练逻辑回归模型...")
    lr_model = LogisticRegression(
        C=1.0,
        max_iter=1000,
        multi_class='multinomial',
        class_weight='balanced',
        random_state=42
    )
    lr_model.fit(X_train_tfidf, y_train)

    y_pred_lr = lr_model.predict(X_test_tfidf)
    lr_accuracy = accuracy_score(y_test, y_pred_lr)
    print(f"逻辑回归准确率: {lr_accuracy:.4f}")

    # 选择最佳模型
    models = {
        'SVM': (svm_model, svm_accuracy, y_pred_svm),
        'RandomForest': (rf_model, rf_accuracy, y_pred_rf),
        'LogisticRegression': (lr_model, lr_accuracy, y_pred_lr)
    }

    best_model_name = max(models.items(), key=lambda x: x[1][1])[0]
    best_model, best_accuracy, best_predictions = models[best_model_name]

    print(f"\n最佳模型: {best_model_name}, 准确率: {best_accuracy:.4f}")

    # 详细分类报告
    print("\n最佳模型分类报告:")
    print(classification_report(y_test, best_predictions, target_names=label_encoder.classes_))

    # 混淆矩阵可视化
    plt.figure(figsize=(12, 10))
    cm = confusion_matrix(y_test, best_predictions)

    # 转换为百分比
    cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

    sns.heatmap(cm_percent, annot=True, fmt='.1f',
                xticklabels=label_encoder.classes_,
                yticklabels=label_encoder.classes_,
                cmap='Blues')
    plt.title(f'混淆矩阵 ({best_model_name})')
    plt.ylabel('真实标签')
    plt.xlabel('预测标签')
    plt.tight_layout()
    plt.savefig('mbti_confusion_matrix.png')

    # 特征重要性 (SVM系数或随机森林特征重要性)
    if best_model_name == 'SVM' or best_model_name == 'LogisticRegression':
        # 获取系数
        if best_model_name == 'SVM':
            coefficients = best_model.coef_
        else:
            coefficients = best_model.coef_

        # 对于每个类别
        for i, mbti_type in enumerate(label_encoder.classes_):
            # 提取前10个最重要特征
            feature_names = vectorizer.get_feature_names_out()
            if i < coefficients.shape[0]:  # 确保有足够的系数行
                top_features_idx = np.argsort(coefficients[i])[-10:]
                top_features = [(feature_names[j], coefficients[i][j]) for j in top_features_idx]

                print(f"\n{mbti_type} 的前10个最重要特征:")
                for feature, coef in top_features:
                    print(f"{feature}: {coef:.4f}")

    elif best_model_name == 'RandomForest':
        feature_importances = best_model.feature_importances_
        feature_names = vectorizer.get_feature_names_out()

        # 前10个最重要特征
        top_features_idx = np.argsort(feature_importances)[-10:]
        top_features = [(feature_names[i], feature_importances[i]) for i in top_features_idx]

        print("\n随机森林前10个最重要特征:")
        for feature, importance in top_features:
            print(f"{feature}: {importance:.4f}")

    # 返回最佳模型及相关组件
    return {
        'best_model': best_model,
        'best_model_name': best_model_name,
        'accuracy': best_accuracy,
        'vectorizer': vectorizer,
        'label_encoder': label_encoder,
        'X_test': X_test_tfidf,
        'y_test': y_test,
        'y_pred': best_predictions
    }

# 4. 预测函数
def predict_mbti(text, model, vectorizer, label_encoder):
    # 预处理文本
    clean_text = preprocess_text(text)

    # 向量化
    X = vectorizer.transform([clean_text])

    # 预测
    pred_idx = model.predict(X)[0]
    pred_mbti = label_encoder.inverse_transform([pred_idx])[0]

    # 如果模型支持predict_proba，获取概率
    if hasattr(model, 'predict_proba'):
        probs = model.predict_proba(X)[0]
        class_probs = {label_encoder.inverse_transform([i])[0]: prob
                      for i, prob in enumerate(probs)}

        # 排序
        sorted_probs = {k: v for k, v in sorted(
            class_probs.items(), key=lambda item: item[1], reverse=True
        )}

        return pred_mbti, sorted_probs
    else:
        # 不支持概率的模型
        decision = model.decision_function(X)
        if decision.ndim > 1:  # 多类决策函数
            scores = decision[0]
            classes = label_encoder.classes_
            class_scores = {classes[i]: scores[i] for i in range(len(classes))}
            sorted_scores = {k: v for k, v in sorted(
                class_scores.items(), key=lambda item: item[1], reverse=True
            )}
            return pred_mbti, sorted_scores
        else:
            return pred_mbti, None

# 直接用当前已经加载的df变量
print(f"数据集形状: {df.shape}")
print(f"列名: {df.columns.tolist()}")
print(f"MBTI类型分布:\n{df['type'].value_counts()}")

# 预处理数据
df, label_encoder = process_data(df)
if df is not None:
    # 构建并评估模型
    model_results = build_models(df, label_encoder)

    # 测试预测函数
    test_texts = [
        "I enjoy spending time alone with books. I prefer planning everything in advance and following a schedule. I need time to process my emotions and often analyze why I feel a certain way.",
        "I love parties and meeting new people! I'm spontaneous and go with the flow. I make decisions based on how I feel and what aligns with my values.",
        "I'm analytical and logical. I enjoy solving complex problems and developing systems. I prefer working independently and dislike unnecessary social interactions."
    ]

    print("\n样本文本预测:")
    for i, text in enumerate(test_texts):
        predicted_mbti, probabilities = predict_mbti(
            text,
            model_results['best_model'],
            model_results['vectorizer'],
            model_results['label_encoder']
        )

        print(f"\n示例 {i+1}: {text[:100]}...")
        print(f"预测的MBTI类型: {predicted_mbti}")

        if probabilities:
            print("各类型概率/分数:")
            for j, (mbti, prob) in enumerate(probabilities.items()):
                if j < 5:  # 只显示前5个
                    print(f"{mbti}: {prob:.4f}")
```

```python
# BERT-MBTI分类器
# 使用预训练的BERT模型进行MBTI人格类型预测

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import re
from collections import Counter
import time
from tqdm import tqdm
import transformers
from transformers import BertTokenizer, BertModel, AdamW, get_linear_schedule_with_warmup

# 设置随机种子以确保结果可重现
SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# 检查GPU可用性
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"使用设备: {device}")

# 1. 文本预处理函数
def preprocess_text(text):
    """清理文本数据"""
    # 删除URL
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # 删除HTML标签
    text = re.sub(r'<.*?>', '', text)
    # 只保留字母和空格
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # 删除多余空格
    text = re.sub(r'\s+', ' ', text)
    # 转换为小写并去除前后空格
    text = text.lower().strip()
    return text

# 2. 自定义MBTI数据集类
class MBTIDataset(Dataset):
    """MBTI数据集类"""
    def __init__(self, texts, labels, tokenizer, max_len):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]

        # 使用BERT tokenizer处理文本
        encoding = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_len,
            padding='max_length',
            truncation=True,
            return_token_type_ids=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        return {
            'text': text,
            'input_ids': encoding['input_ids'].flatten(),
            'attention_mask': encoding['attention_mask'].flatten(),
            'token_type_ids': encoding['token_type_ids'].flatten(),
            'label': torch.tensor(label, dtype=torch.long)
        }

# 3. 数据准备函数
def prepare_data(df, bert_model_name='bert-base-uncased', max_len=128, augment=True):
    """准备数据用于BERT模型"""
    print("开始数据预处理...")

    # 检查是否已有处理好的列，如果没有则处理
    if 'combined_clean_posts' not in df.columns:
        # 分割帖子
        print("分割帖子...")
        if not isinstance(df["posts"].iloc[0], list):
            df["posts"] = df["posts"].apply(lambda x: x.split("|||"))

        # 清理文本
        print("清理文本...")
        df["clean_posts"] = df["posts"].apply(lambda posts: [preprocess_text(p) for p in posts])

        # 合并清理后的帖子为一个文本
        print("合并帖子...")
        df["combined_clean_posts"] = df["clean_posts"].apply(lambda posts: " ".join(posts))

    # 由于BERT有长度限制，可能需要截断长文本
    print(f"截断文本至最大长度 {max_len*10} 个字符...")
    df["truncated_text"] = df["combined_clean_posts"].apply(lambda x: x[:max_len*10])

    # 编码标签
    print("编码MBTI类型...")
    label_encoder = LabelEncoder()
    df["label"] = label_encoder.fit_transform(df["type"])

    # 显示标签分布
    label_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
    print(f"标签映射: {label_mapping}")
    label_counts = Counter(df["label"])
    print(f"标签分布: {label_counts}")

    # 分割训练集和测试集
    train_df, test_df = train_test_split(
        df, test_size=0.2, random_state=SEED, stratify=df["label"]
    )

    # 数据增强 - 为少数类复制样本
    if augment:
        print("应用数据增强...")
        # 计算每个类别需要增强的样本数量
        class_counts = Counter(train_df["label"])
        max_count = max(class_counts.values())
        target_count = int(max_count * 0.5)  # 至少为最大类别的50%

        augmented_dfs = []

        # 对少数类进行过采样
        for label, count in class_counts.items():
            if count < target_count:
                # 获取该类别的样本
                class_samples = train_df[train_df["label"] == label]
                # 需要复制的倍数
                multiplier = target_count // count + 1
                # 复制样本
                for _ in range(multiplier):
                    if len(augmented_dfs) * count + len(class_samples) <= target_count:
                        augmented_dfs.append(class_samples)

        # 合并增强后的数据
        if augmented_dfs:
            augment_df = pd.concat(augmented_dfs, ignore_index=True)
            train_df = pd.concat([train_df, augment_df], ignore_index=True)
            print(f"通过数据增强添加了 {len(augment_df)} 个样本")
            print(f"增强后的训练集分布: {Counter(train_df['label'])}")

    # 加载BERT tokenizer
    print(f"加载BERT tokenizer: {bert_model_name}")
    tokenizer = BertTokenizer.from_pretrained(bert_model_name)

    # 创建数据集
    train_dataset = MBTIDataset(
        texts=train_df["truncated_text"].values,
        labels=train_df["label"].values,
        tokenizer=tokenizer,
        max_len=max_len
    )

    test_dataset = MBTIDataset(
        texts=test_df["truncated_text"].values,
        labels=test_df["label"].values,
        tokenizer=tokenizer,
        max_len=max_len
    )

    print(f"训练集大小: {len(train_dataset)}, 测试集大小: {len(test_dataset)}")

    return train_dataset, test_dataset, label_encoder, tokenizer

# 4. BERT分类模型
class BERTMBTIClassifier(nn.Module):
    """使用BERT的MBTI分类器"""
    def __init__(self, bert_model_name, num_classes, dropout=0.3):
        super(BERTMBTIClassifier, self).__init__()

        # 加载预训练的BERT模型
        self.bert = BertModel.from_pretrained(bert_model_name)

        # BERT输出的隐藏状态维度
        hidden_size = self.bert.config.hidden_size

        # 分类层
        self.dropout = nn.Dropout(dropout)
        self.fc1 = nn.Linear(hidden_size, hidden_size // 2)
        self.fc2 = nn.Linear(hidden_size // 2, num_classes)
        self.relu = nn.ReLU()

    def forward(self, input_ids, attention_mask, token_type_ids):
        # BERT前向传播
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids
        )

        # 获取[CLS]标记的最后隐藏状态
        pooled_output = outputs.pooler_output

        # 分类层
        x = self.dropout(pooled_output)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)

        return x

# 5. 训练函数
def train_bert_model(model, train_loader, val_loader, optimizer, scheduler, criterion, epochs=4):
    """训练BERT模型"""
    best_val_loss = float('inf')
    best_val_acc = 0.0
    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []

    for epoch in range(epochs):
        print(f'\n{"-" * 20} Epoch {epoch+1}/{epochs} {"-" * 20}')

        # 训练模式
        model.train()
        train_loss = 0
        train_acc = 0

        # 训练循环
        for batch in tqdm(train_loader, desc="Training"):
            # 获取批次数据
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            labels = batch['label'].to(device)

            # 梯度清零
            optimizer.zero_grad()

            # 前向传播
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                token_type_ids=token_type_ids
            )

            # 计算损失
            loss = criterion(outputs, labels)

            # 反向传播
            loss.backward()

            # 梯度裁剪
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            # 参数更新
            optimizer.step()
            scheduler.step()

            # 计算准确率
            _, preds = torch.max(outputs, dim=1)
            correct = (preds == labels).float().sum()
            acc = correct / len(labels)

            train_loss += loss.item()
            train_acc += acc.item()

        # 计算平均训练损失和准确率
        avg_train_loss = train_loss / len(train_loader)
        avg_train_acc = train_acc / len(train_loader)
        train_losses.append(avg_train_loss)
        train_accs.append(avg_train_acc)

        # 验证模式
        model.eval()
        val_loss = 0
        val_acc = 0

        with torch.no_grad():
            for batch in tqdm(val_loader, desc="Validation"):
                # 获取批次数据
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                token_type_ids = batch['token_type_ids'].to(device)
                labels = batch['label'].to(device)

                # 前向传播
                outputs = model(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    token_type_ids=token_type_ids
                )

                # 计算损失
                loss = criterion(outputs, labels)

                # 计算准确率
                _, preds = torch.max(outputs, dim=1)
                correct = (preds == labels).float().sum()
                acc = correct / len(labels)

                val_loss += loss.item()
                val_acc += acc.item()

        # 计算平均验证损失和准确率
        avg_val_loss = val_loss / len(val_loader)
        avg_val_acc = val_acc / len(val_loader)
        val_losses.append(avg_val_loss)
        val_accs.append(avg_val_acc)

        # 打印结果
        print(f'Train Loss: {avg_train_loss:.4f} | Train Acc: {avg_train_acc*100:.2f}%')
        print(f'Val. Loss: {avg_val_loss:.4f} | Val. Acc: {avg_val_acc*100:.2f}%')

        # 保存最佳模型
        if avg_val_acc > best_val_acc:
            best_val_acc = avg_val_acc
            torch.save(model.state_dict(), 'best-bert-mbti-model.pt')
            print(f"[!] 已保存最佳模型 (验证准确率: {best_val_acc*100:.2f}%)")

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            torch.save(model.state_dict(), 'best-bert-mbti-model-loss.pt')
            print(f"[!] 已保存最佳模型 (验证损失: {best_val_loss:.4f})")

    # 绘制训练过程
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(train_accs, label='Train Accuracy')
    plt.plot(val_accs, label='Val Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.tight_layout()
    plt.savefig('bert_mbti_training.png')
    plt.show()

    return train_losses, val_losses, train_accs, val_accs

# 6. 评估函数
def evaluate_bert_model(model, test_loader, label_encoder):
    """评估BERT模型性能"""
    model.eval()
    predictions = []
    actual_labels = []

    with torch.no_grad():
        for batch in tqdm(test_loader, desc="Evaluating"):
            # 获取批次数据
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            token_type_ids = batch['token_type_ids'].to(device)
            labels = batch['label'].to(device)

            # 前向传播
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                token_type_ids=token_type_ids
            )

            # 获取预测结果
            _, preds = torch.max(outputs, dim=1)

            predictions.extend(preds.cpu().tolist())
            actual_labels.extend(labels.cpu().tolist())

    # 计算准确率
    accuracy = accuracy_score(actual_labels, predictions)

    # 生成分类报告
    class_names = label_encoder.classes_
    report = classification_report(actual_labels, predictions, target_names=class_names)

    # 混淆矩阵
    plt.figure(figsize=(12, 10))
    cm = confusion_matrix(actual_labels, predictions)
    cm_percent = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100

    sns.heatmap(cm_percent, annot=True, fmt='.1f',
                xticklabels=class_names,
                yticklabels=class_names,
                cmap='Blues')
    plt.title('Confusion Matrix (Percent)')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout()
    plt.savefig('bert_mbti_confusion_matrix.png')

    return accuracy, report

# 7. 预测函数
def predict_with_bert(text, model, tokenizer, label_encoder, max_len=128):
    """使用BERT模型预测单个文本的MBTI类型"""
    model.eval()

    # 文本预处理
    text = preprocess_text(text)

    # 使用BERT tokenizer
    encoding = tokenizer.encode_plus(
        text,
        add_special_tokens=True,
        max_length=max_len,
        padding='max_length',
        truncation=True,
        return_token_type_ids=True,
        return_attention_mask=True,
        return_tensors='pt'
    )

    # 移动张量到设备
    input_ids = encoding['input_ids'].to(device)
    attention_mask = encoding['attention_mask'].to(device)
    token_type_ids = encoding['token_type_ids'].to(device)

    # 预测
    with torch.no_grad():
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            token_type_ids=token_type_ids
        )

        # 获取预测和概率
        probabilities = torch.softmax(outputs, dim=1)[0]
        predicted_class = torch.argmax(probabilities).item()

    # 转换预测结果为MBTI类型
    predicted_mbti = label_encoder.inverse_transform([predicted_class])[0]

    # 获取所有类型的概率
    all_probs = {label_encoder.inverse_transform([i])[0]: prob.item()
                for i, prob in enumerate(probabilities)}

    # 排序概率
    sorted_probs = {k: v for k, v in sorted(all_probs.items(), key=lambda item: item[1], reverse=True)}

    return predicted_mbti, sorted_probs

# 8. 主函数 - 运行BERT模型
def run_bert_model(df, bert_model_name='bert-base-uncased', max_len=128, batch_size=16, epochs=4):
    """运行BERT模型进行MBTI预测"""
    print(f"使用BERT模型: {bert_model_name}")

    # 准备数据
    train_dataset, test_dataset, label_encoder, tokenizer = prepare_data(
        df, bert_model_name=bert_model_name, max_len=max_len
    )

    # 创建加权随机采样器来处理类别不平衡
    train_labels = [sample['label'].item() for sample in train_dataset]
    class_counts = Counter(train_labels)
    class_weights = {class_id: 1.0 / count for class_id, count in class_counts.items()}
    sample_weights = [class_weights[label] for label in train_labels]
    sampler = WeightedRandomSampler(
        weights=sample_weights,
        num_samples=len(train_dataset),
        replacement=True
    )

    # 创建数据加载器
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        sampler=sampler
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size
    )

    # 初始化模型
    num_classes = len(label_encoder.classes_)
    model = BERTMBTIClassifier(bert_model_name, num_classes)
    model = model.to(device)

    # 打印模型架构
    print(f"模型架构:\n{model}")

    # 定义优化器、调度器和损失函数
    # 使用不同的学习率
    param_optimizer = list(model.named_parameters())
    no_decay = ['bias', 'LayerNorm.bias', 'LayerNorm.weight']
    optimizer_grouped_parameters = [
        {'params': [p for n, p in param_optimizer if not any(nd in n for nd in no_decay)], 'weight_decay': 0.01},
        {'params': [p for n, p in param_optimizer if any(nd in n for nd in no_decay)], 'weight_decay': 0.0}
    ]

    optimizer = AdamW(optimizer_grouped_parameters, lr=2e-5)

    # 计算总训练步数
    total_steps = len(train_loader) * epochs

    # 创建学习率调度器
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=0,
        num_training_steps=total_steps
    )

    # 计算类别权重
    class_counts = Counter(train_labels)
    weight_values = torch.FloatTensor([
        1.0 / (class_counts[i] / len(train_labels))
        for i in range(num_classes)
    ]).to(device)

    # 定义损失函数
    criterion = nn.CrossEntropyLoss(weight=weight_values)

    # 训练模型
    print("\n开始训练BERT模型...")
    train_bert_model(model, train_loader, test_loader, optimizer, scheduler, criterion, epochs=epochs)

    # 加载最佳模型
    model.load_state_dict(torch.load('best-bert-mbti-model.pt'))

    # 评估模型
    print("\n评估最佳模型...")
    accuracy, report = evaluate_bert_model(model, test_loader, label_encoder)

    print(f"\n测试集准确率: {accuracy*100:.2f}%")
    print(f"\n分类报告:\n{report}")

    # 样本预测
    test_texts = [
        "I enjoy spending time alone with books. I prefer planning everything in advance and following a schedule. I need time to process my emotions and often analyze why I feel a certain way.",
        "I love parties and meeting new people! I'm spontaneous and go with the flow. I make decisions based on how I feel and what aligns with my values.",
        "I'm analytical and logical. I enjoy solving complex problems and developing systems. I prefer working independently and dislike unnecessary social interactions."
    ]

    print("\n样本文本预测:")
    for i, text in enumerate(test_texts):
        mbti_type, probabilities = predict_with_bert(text, model, tokenizer, label_encoder, max_len)

        print(f"\n示例 {i+1}: {text[:100]}...")
        print(f"预测的MBTI类型: {mbti_type}")
        print("各类型概率:")
        for j, (mbti, prob) in enumerate(probabilities.items()):
            if j < 5:  # 只显示前5个
                print(f"{mbti}: {prob:.4f}")

    return {
        'model': model,
        'tokenizer': tokenizer,
        'label_encoder': label_encoder,
        'accuracy': accuracy,
        'report': report
    }

# 主程序
if __name__ == "__main__":
    # 使用已加载的df数据框
    print(f"数据集形状: {df.shape}")

    # 运行BERT模型
    results = run_bert_model(
        df,
        bert_model_name='bert-base-uncased',  # 也可以使用其他BERT变体如'bert-large-uncased'
        max_len=128,                          # BERT序列最大长度
        batch_size=16,                        # 根据GPU内存调整
        epochs=4                              # BERT通常只需要几个epoch
    )

    print("\nBERT模型训练和评估完成!")
```
